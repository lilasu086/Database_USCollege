###**5.1.1 Average SAT scores according to States**

**Aim:**

Exploring SAT scores across states, we aim to determine the average scores for each state and identify which state boasts the highest average SAT score.

```python
%%bigquery --project $project_id
SELECT
  state,region,
  ROUND(AVG(med_sat_value)) AS mean_sat,
  RANK() OVER (ORDER BY AVG(med_sat_value) DESC) AS ranking,
  ROUND(AVG(grad_100_value), 2) AS avg_grad100_rate,
  ROUND(AVG(grad_150_value), 2) AS avg_grad150_rate
FROM
  `ba-775-a06.cc_datasets.1-details`
WHERE
  med_sat_value IS NOT NULL AND grad_100_value IS NOT NULL AND grad_150_value IS NOT NULL
GROUP BY
  state,region
ORDER BY
  ranking;
```

###**5.2.11 Exploring the Correlation**

**Aim:**

Examine the correlation between different SAT score groups and their average graduation rates over multiple years.

```python
%%bigquery --project $project_id
SELECT
  CASE
  WHEN med_sat_value BETWEEN 1350 AND 1600 THEN 'High_SAT_Score'
   WHEN med_sat_value BETWEEN 1200 AND 1350 THEN 'Higher_Median_SAT_Score'
   WHEN med_sat_value BETWEEN 1050 AND 1200 THEN 'Lower_Median_SAT_Score'
   WHEN med_sat_value BETWEEN 0 AND 1050 THEN 'Low_SAT_Score'
   END AS sat, A.year,ROUND(AVG(A.grad_100_rate),2) AS avg_grad100_rate,
   ROUND(AVG(A.grad_150_rate),2) AS avg_grad150_rate
FROM `ba-775-a06.cc_datasets.2-grads` AS A
LEFT JOIN `ba-775-a06.cc_datasets.1-details` AS B
USING (unitid)
WHERE B.med_sat_value IS NOT NULL AND A.grad_100_rate IS NOT NULL AND A.grad_150_rate IS NOT NULL
GROUP BY sat,year
ORDER BY CASE
    WHEN sat = 'High_SAT_Score' THEN 1
    WHEN sat = 'Higher_Median_SAT_Score' THEN 2
    WHEN sat = 'Lower_Median_SAT_Score' THEN 3
    WHEN sat = 'Low_SAT_Score' THEN 4
  END,year
```

**Insights:**

The analysis uncovers a noticeable upward trend in graduation rates over the years. Additionally, a significant correlation emerges between Median SAT scores and graduation rates, indicating that higher SAT score groups consistently achieve better graduation outcomes. This finding implies a positive association between academic preparedness, as gauged by SAT scores, and successful program completion throughout the examined time frame.

###**5.3.2 Comparison between SAT scores and Flagship Colleges**

```python
%%bigquery --project $project_id
WITH CollegeRanks AS (
   SELECT
       state,
       chronname,
       med_sat_value,
       RANK() OVER (PARTITION BY state ORDER BY med_sat_value DESC) AS sat_rank
   FROM
       `ba-775-a06.cc_datasets.1-details`
   WHERE
       med_sat_value IS NOT NULL  -- Exclude NULL values if necessary
)


SELECT
   t.state,
   t.chronname AS flagship_college,
   t.med_sat_value AS flagship_sat_value,
   r.chronname AS college_with_highest_sat,
   r.med_sat_value AS highest_sat_value
FROM
   `ba-775-a06.cc_datasets.1-details` t
JOIN
   CollegeRanks r ON t.state = r.state AND r.sat_rank = 1
WHERE
   t.flagship = 'X'  -- Filter for flagship colleges
   AND t.med_sat_value IS NOT NULL
ORDER BY
   flagship_sat_value DESC,t.state, t.chronname;
```

**Insights:**

The obtained results shed light on an intriguing aspect: the SAT score does not necessarily serve as a definitive indicator for a university to be classified as a flagship institution. A compelling case for comparison is the state of Massachusetts, where the flagship university, UMASS Amherst, boasts a SAT score of 1209. In contrast, the highest-ranking college in the state, the Massachusetts Institute of Technology (MIT), significantly surpasses this score with an impressive 1503. This disparity challenges the assumption that a high SAT score is a requisite criterion for a university to attain flagship status. It underscores the complexity of factors influencing such classifications, urging a more nuanced examination beyond standardized test scores to truly discern the distinctive qualities that contribute to a university being designated as a flagship institution.

So far SAT score did not come out to be the unique indicator for the flagship universities, we'll now look upon the average awards with respect to other colleges and flagship universties in the state, answering the following question -

###**5.3.3 Comparison between Awards and Flagship Colleges**

```python
%%bigquery --project $project_id

WITH FlagshipComparison AS (
    SELECT
        state,
        AVG(CASE WHEN flagship = 'X' THEN awards_per_value END) AS flagship_awards
    FROM
        `ba-775-a06.cc_datasets.1-details`
    WHERE
        awards_per_value IS NOT NULL
    GROUP BY
        state
)

SELECT
    t.state,
    AVG(t.awards_per_value) AS avg_awards_all_colleges,
    fc.flagship_awards AS awards_flagship,
    fc.flagship_awards - AVG(t.awards_per_value) AS awards_difference
FROM
    `ba-775-a06.cc_datasets.1-details` t
JOIN
    FlagshipComparison fc ON t.state = fc.state
WHERE
    t.awards_per_value IS NOT NULL AND fc.flagship_awards IS NOT NULL
GROUP BY
    t.state, fc.flagship_awards
ORDER BY
    awards_flagship DESC, t.state;
```

**Insights**

The table reveals a discernible pattern that could offer valuable insights in favor of flagship universities over others. Surprisingly, the awards value for the flagship university exceeded the average awards value across the state. This observation is inherently explanatory, considering that flagship universities are among the oldest and are typically state run public institutions.
